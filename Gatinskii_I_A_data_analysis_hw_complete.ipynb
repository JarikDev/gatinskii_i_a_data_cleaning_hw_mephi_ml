1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:

Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

```python
#ваш код здесь
display(hh_df['Образование и ВУЗ'][:5])
def extract_education(education: str):
    educations = ['высшее','неоконченное высшее','среднее специальное','среднее']
    education = education.lower()
    for e in educations:
        if education.startswith(e):
            return e
    else:
        return 'неизвестно'
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(lambda x: extract_education(x))
hh_df = hh_df.drop(columns=['Образование и ВУЗ'])
display(hh_df.head())
display(hh_df['Образование'].value_counts())
```

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина.

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь

def get_work_experience_in_months(input):
    input = str(input).lower()
    if input in ['nan', 'не указано']:
        return np.nan
    chunks = input.split(" ")
    years:int = 0
    months:int = 0
    for i, chunk in enumerate(chunks):
        if chunk.isnumeric():
            if 'мес' in chunks[i + 1]:
                months = int(chunk)
                break
            elif years == 0 and ('год' in chunks[i + 1]  or 'лет' in chunks[i + 1]):
                years = int(chunk)
                if 'мес' not in input:
                    break      
    return years * 12 + months

hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_work_experience_in_months)
hh_df['Опыт работы (месяц)'].median()
hh_df = hh_df.drop(columns=['Опыт работы'])
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']

def get_city(line):
    if line is np.nan:
        return "другие"
    city = line.split(",")[0].strip()
    if city in million_cities:
        return "город-миллионник"
    elif city in ["Москва","Санкт-Петербург"]:
        return city
    else:
        return "другие"


def is_ready_for_activity(line, activity:str):
    if line is np.nan:
        return False
    chunks = line.split(",")
    for c in chunks:
        if activity in c:
            if "не " in c:
                return False 
            else:
                return True
    else:
        return False 

hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(get_city)
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(lambda x: is_ready_for_activity(x,"перее"))
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(lambda x: is_ready_for_activity(x,"командировка"))
display(hh_df['Город'].value_counts(normalize=True))
is_ready_for_reloc = (hh_df['Готовность к переезду'] == True)
is_ready_for_trip = (hh_df['Готовность к командировкам'] == True)
ready_for_activities = hh_df[is_ready_for_reloc & is_ready_for_trip].shape[0]
display(f"готово к переездам и командировкам {round((ready_for_activities / hh_df.shape[0])*100)}")
hh_df = hh_df.drop(columns=['Город, переезд, командировки'])
```

```python
#ваш код здесь
ex_rate_df: pd.DataFrame = pd.read_csv("D:\prog\proj\python\mifi_ml\mifi_ml\python_for_data_analysis\data\ExchangeRates.csv")
ex_rate_df['date'] = pd.to_datetime(ex_rate_df['date'], format='%d/%m/%y')

mapper_dict = {'руб.':'RUB','грн.':'UAH','USD':'USD','EUR':'EUR','бел.руб.':'BYN','KGS':'KGS','сум':'UZS','AZN':'AZN','KZT':'KZT'}

hh_df['Обновление резюме'] = hh_df['Обновление резюме'].apply(lambda x: x.split(" ")[0])
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], format='%d.%m.%Y')

def to_iso_name(line):
    val = line.split(" ")[1]
    if val in mapper_dict:
        return mapper_dict[val]
    else:
        return "UNKNOWN"

hh_df['валюта'] = hh_df['ЗП'].apply(to_iso_name)
hh_df['ЗП'] = hh_df['ЗП'].apply(lambda x: int(x.split(" ")[0]))
hh_df = pd.merge(hh_df,ex_rate_df[['date', 'currency','close','proportion']], left_on=['Обновление резюме', 'валюта'], right_on=['date', 'currency'], how='left')
hh_df['close'] = hh_df.apply(lambda x: 1 if x['валюта'] =='RUB' else x['close'], axis=1)
hh_df['proportion'] = hh_df.apply(lambda x: 1 if x['валюта'] =='RUB' else x['proportion'], axis=1)
hh_df['ЗП (руб)'] = hh_df['ЗП']*hh_df['close']/ hh_df['proportion']
hh_df = hh_df.drop(columns=['ЗП','date', 'currency','close','proportion','валюта'])
display(hh_df['ЗП (руб)'].median())
display(hh_df.info())
```

```python
# ваш код здесь
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

data = hh_df['Возраст']
mu = data.mean()
sigma = data.std()
lower_bound = mu - 3 * sigma
upper_bound = mu + 4 * sigma 
display(f"В категорию отклоненией попадают соискатели моложе {lower_bound} и старше {upper_bound}")
outliers, cleaned = outliers_z_score(hh_df, 'Возраст', log_scale=True)
display(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
display(f'Результирующее число записей: {cleaned.shape[0]}')

log_data = np.log(hh_df['Возраст'])
mu = log_data.mean()
sigma = log_data.std()
lower_bound = mu - 3 * sigma
upper_bound = mu + 4 * sigma
ax = sns.histplot(log_data, log=True)
ax.set_title('Распределение признака в Возраст в логарифмическом масштабе')
ax.set_ylabel('Количество')
ax.axvline(log_data.mean(), color='k', lw=2)
ax.axvline(lower_bound, color='k', lw=2)
ax.axvline(upper_bound, color='k', lw=2)

# очищенная выборка
hh_df = cleaned
display(hh_df.shape[0])
```
